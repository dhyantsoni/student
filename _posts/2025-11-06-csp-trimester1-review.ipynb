As we wrap up Trimester 1 in Computer Science Principles, it's valuable to reflect on our journey and the key milestones we've achieved. This trimester has been divided into three major components, each building upon the previous to create a solid foundation in computer science.

## Sprint 1: Setting Up Our Development Environment

Our first sprint focused on establishing a robust development environment, which is crucial for any programmer. Key accomplishments included:

- Installing and configuring essential development tools
- Setting up GitHub and learning version control basics
- Configuring VS Code with necessary extensions
- Understanding the importance of proper tool configuration
- Learning to use terminal commands effectively

I enjoyed making a calculator lesson and doing OOP breakout, along with the other background projects

This foundation was critical as it set us up for success in all future programming endeavors.

## Sprint 2: Diving into JavaScript and Python

The second sprint expanded our programming horizons by introducing us to two powerful programming languages: **Javascript** and **Python**

Sprint 2 focused on applying programming fundamentals through both JavaScript and Python. While I was already familiar with most of the concepts covered such as variables, conditionals, loops, and algorithms this sprint helped me refine my approach and improve how I structure and think about code.

### Key Reflections:
- **Prior Knowledge**: I already understood most of the concepts, so this sprint was more about deepening my understanding rather than learning from scratch.  
- **Lucas’s Lab**: This was one of the most challenging parts of the sprint. Even though I knew the material, the lab pushed me to think critically, apply logic efficiently, and handle complex layers of conditions and iteration.  
- **JavaScript and Python Practice**: Switching between the two languages helped me focus on logic instead of syntax, improving my ability to adapt and translate solutions between languages.  
- **Algorithmic Thinking**: Reinforced the habit of breaking problems down into smaller, manageable steps before coding a skill that proved useful for debugging and optimization.  
- **Growth Through Challenge**: Even when the concepts were familiar, the assignments and labs made me realize there’s always room to improve in efficiency, clarity, and code design.
